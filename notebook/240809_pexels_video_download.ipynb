```python
import subprocess
from pathlib import Path

def download_video_id(video_id, remove_if_exists=False) -> bool:
    output_path = f"../data/pexels-videos/{video_id}.mp4"
    if Path(output_path).exists():
        if remove_if_exists:
            Path(output_path).unlink()
            print(f"Removed existing file {output_path}.")
        else:
            print(f"File {output_path} already exists.")
            return True
    system_command = f'wget -O "{output_path}" $(curl -Ls -o /dev/null -w %{{url_effective}} "https://www.pexels.com/download/video/{video_id}/")'
    print(system_command)
    # run command, don't let output print to console, and return True if successful.
    result = subprocess.run(system_command, shell=True, stdout=subprocess.PIPE, stderr=subprocess.PIPE)
    if result.returncode == 0:
        print(f"Downloaded video {video_id} to {output_path}.")
        return True
    else:
        print(f"Failed to download video {video_id}.")
        # print console.
        print(result.stdout.decode())
        print(result.stderr.decode())
        return False

# test
"Success" if download_video_id(
    df["pexels_video_id"].iloc[0],
    remove_if_exists=True,
) else "Failed"
```

```python
# find all mp4 files in "../data/pexels-videos/*",
# crop them to square,
# and resize them to 512x512.

import subprocess
import cv2

def get_video_width_height(video_path) -> tuple:
    # Open the video file
    cap = cv2.VideoCapture(str(video_path))
    # Get the width and height of the video
    width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    cap.release()
    return width, height

def crop_and_resize(input_video, output_video, out_size=128, overwrite=False) -> bool:
    if Path(output_video).exists():
        if not overwrite:
            print(f"File {output_video} already exists.")
            return True
        else:
            Path(output_video).unlink()
            print(f"Removed existing file {output_video}.")
    # Crop the center square and resize to 512x512 using ffmpeg
    crop_size = min(get_video_width_height(input_video))
    result = subprocess.run([
        'ffmpeg', 
        '-i', input_video, 
        '-vf', f'crop={crop_size}:{crop_size},scale={out_size}:{out_size}',
        '-c:a', 'copy', 
        output_video
    ], stdout=subprocess.PIPE, stderr=subprocess.PIPE)
    if result.returncode == 0:
        print(f"Resized video {input_video} to {output_video}.")
        return True
    else:
        print(result.stdout.decode())
        print(result.stderr.decode())
        raise Exception(f"Failed to resize video {input_video}.")

paths = list(Path("../data/pexels-videos").glob("*.mp4"))
print(len(paths))
# remove all paths ending with "-small.mp4".
paths = [p for p in paths if not p.name.endswith("-small.mp4")]
print(len(paths))

input_video = paths[0]
output_video = paths[0].parent / f"{paths[0].stem}-small.mp4"
print(input_video, '->', output_video)

crop_and_resize(input_video, output_video, overwrite=True)

```

```python
for path in tqdm(paths):
    output_video = path.parent / f"{path.stem}-small.mp4"
    crop_and_resize(path, output_video, out_size=128, overwrite=False)
```
